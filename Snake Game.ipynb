# Snake:

### General function of the game

Snake is a game where a snake moves through a plane in 2D and eats food to grow.
Once it consumes a unit of food it grows a unit.
When the snake collides with itself the game is over and restarts.

### Pseudocode

Print a board width w and height h.
Generate random food position, iterate once snake eats.
Define snake position
Define snake directions input by user up,down,left,right
Define food collision

```python
#Import needed libraries
import turtle
import random
import pygame
from pygame.locals import *
from pygame import mixer


#Music for the game :DDD
#mixer.init()
#mixer.music.load('bensound-summer_ogg_music.ogg')
#mixer.music.play()


#Screen variables
w = 800
h = 800
food_size = 20
delay=30


#Offsets for new part of snake once food eated
offsets = {
    "up": (0, 20),
    "down": (0, -20),
    "left": (-20, 0),
    "right": (20, 0)
}


#Define initial values
def initial():
    global food_pos, pen, snake, snake_dir
    #snake is generated
    snake = [[0, 0], [0, 20], [0, 40], [0, 60], [0, 80]]
    snake_dir = "up"
    #initial position food
    food_pos = random_food_pos()
    food.goto(food_pos)
    snake_direction()

    
#Define snake movement
def snake_direction():
    global snake_dir
    
    #snake's head movement
    head = snake[-1].copy()
    
    head[0] = snake[-1][0] + offsets[snake_dir][0]
    head[1] = snake[-1][1] + offsets[snake_dir][1]
    
    #snake collition with itself 
    if head in snake[:-1]:
        initial()

    #snake does not collide
    else:
        snake.append(head)
 
    #snake's old tail is removed 
        if not eat_food():
            snake.pop(0)
 
    #snake move to the other side of the screen
        if snake[-1][0] > w / 2:
            snake[-1][0] -= w
        elif snake[-1][0] < - w / 2:
            snake[-1][0] += w
        elif snake[-1][1] > h / 2:
            snake[-1][1] -= h
        elif snake[-1][1] < -h / 2:
            snake[-1][1] += h
 
        #delete pen stamp from previous call
        pen.clearstamps()
 
        #print snake with new head and without old tail in new call
        for segment in snake:
            pen.goto(segment[0], segment[1])
            pen.stamp()

        #show on screen new print 
        turtle.Screen().update()
        #repeat once delays ms happen
        turtle.ontimer(snake_direction, delay)

#Snake eats food, based on small distance returned by distance food + small offset

def eat_food():
    global food_pos
    if distance_food(snake[-1], food_pos) < 25:
        
        food_pos = random_food_pos()
        food.goto(food_pos)
        return True
    return False

#get food position as a point inside the board. Value between half of
#the board width and height-food size 

def random_food_pos():
    x = random.randint(- w / 2 + food_size, w / 2 - food_size)
    y = random.randint(- h / 2 + food_size, h / 2 - food_size)
    return (x, y)

#Function to determine distance between food and snake head

def distance_food(pos1, pos2):
    x1, y1 = pos1
    x2, y2 = pos2

    distance = ((y2-y1)**2+(x2-x1)**2)**0.5
    
    return distance


#Functions for input movement   

def up():
    global snake_dir
    if snake_dir != "down":
        snake_dir = "up"
 
def right():
    global snake_dir
    if snake_dir != "left":
        snake_dir = "right"
 
def down():
    global snake_dir
    if snake_dir!= "up":
        snake_dir = "down"
 
def left():
    global snake_dir
    if snake_dir != "right":
        snake_dir = "left"
 
#screen setup 
turtle.Screen().title("MySnake")
turtle.Screen().bgcolor(random.choice(["green","blue","grey"]))
turtle.Screen().setup(w, h)
turtle.Screen().setup(800, 800)
turtle.Screen().tracer(0)
 
#shape of snake
pen = turtle.Turtle("square")
pen.color(random.choice(["black","white"]))
pen.penup()




#Display of food
food = turtle.Turtle()
food.shape("square")
food.color(random.choice(["yellow","red","black"]))
food.shapesize(food_size/12)
food.penup()
 
#call directions input onkey
turtle.Screen().listen()
turtle.Screen().onkey(up, "Up")
turtle.Screen().onkey(right, "Right")
turtle.Screen().onkey(down, "Down")
turtle.Screen().onkey(left, "Left")
 
#call initial function 
initial()

#Tkinter looping final function to loop everything
turtle.done()
```
